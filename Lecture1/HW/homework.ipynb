````python
# Problem 6: Bonus: Basic Calculator Function
# Define a function called `calculator` that:
# - Takes two numbers and a mathematical operator (string: '+', '-', '*', '/') as parameters
# - Returns the result of the operation on the two numbers
# - Prints an error message if the operator is not recognized
# Test your function with example inputs.
````

````python
def calculator(num1, num2, operator):
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        if num2 != 0:
            return num1 / num2    
        return 0
    else:
        return "üö® Operator not recognized. Please use '+', '-', '*', or '/'."
````

````python
# printing the function call because we are not printing the values on function call
# we are returning values
print(calculator(4,12,'/'))
print(calculator(4,5,'*'))
print(calculator(4,5,'+'))
print(calculator(4,5,'-'))
````

````python
# Problem 10: Simple Interest Calculator
# Write a program that:
# - Takes principal amount, rate of interest (annual), and time (in years) as input from the
# user.
# - Calculates simple interest using the formula: `SI = (P * R * T)/100`
# - Prints the calculated interest.
````

````python
try:
  principle_amount = float(input("Enter the principal amount (P): "))
  rate_of_interest = float(input("Enter the rate of interest (R) in percentage: "))
  time_in_years = float(input("Enter the time (T) in years: "))
  simple_interest = (principle_amount * rate_of_interest * time_in_years) / 100
  print(f"The simple interest is: {simple_interest}")
except ValueError:
  print("üö® Ahhh!!! Gotchaa! Please write a number")
````

````python
# Problem 11: Circle Properties
# Write a program that:
# - Asks the user for the radius of a circle (float).
# - Calculates and prints the circumference and area of the circle.
# - Use `3.1416` for œÄ.
````

````python
try:
  radius = float(input("Let's see how big of a circle you can draw, enter the radius: "))
  pi = 3.1416
  circumference = 2 * pi * radius
  area = pi * (radius ** 2)
  print(f"The circumference of the circle is: {circumference}")
  print(f"The area of the circle is: {area}")
except ValueError:
  print("üö® Ahhh!!! Gotchaa! Please write a number")
````
