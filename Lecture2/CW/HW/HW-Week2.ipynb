## Problem 1: Even or Odd with Loop and Else

```python
# Problem 1: Even or Odd with Loop and Else
# Uses a while loop to repeatedly ask for integers.
# Stops on a negative number; else clause fires when loop ends normally.
flag = True
while flag:
    num = int(input("Enter an integer (negative to stop): "))

    if num < 0:
        # Negative number entered — exit loop without triggering else
        flag = False

    # Use modulus operator to determine even or odd
    if num % 2 == 0:
        print(f"{num} is Even")
    else:
        print(f"{num} is Odd")
else:
    # This block runs only if the loop was NOT exited via break
    print("Input process ended normally.")
```

## Problem 2: Find Prime Numbers in a Range

```python
# Problem 2: Find Prime Numbers in a Range
# Uses break to exit the inner loop when a divisor is found.
# The else clause on the inner loop confirms primality.

start = int(input("Enter start of range: "))
end   = int(input("Enter end of range: "))

print(f"Prime numbers between {start} and {end}:")

for number in range(start, end + 1):
    if number < 2:
        continue  # Numbers less than 2 are not prime

    # Check divisibility from 2 up to sqrt(number)
    for divisor in range(2, int(number ** 0.5) + 1):
        if number % divisor == 0:
            # A divisor was found — not prime, exit inner loop
            break
    else:
        # Inner loop completed without finding a divisor → number is prime
        print(f"{number} is PRIME")
```

## Problem 3: Loop Control Practice

```python
# Problem 3: Loop Control Practice
# Iterates 1-20; skips multiples of 3 with continue;
# breaks if number > 15; else clause detects clean completion.

for i in range(1, 21):
    if i > 15:
        # Stop completely when number exceeds 15
        print(f"Reached {i} which is > 15. Breaking loop.")
        break

    if i % 3 == 0:
        # Skip multiples of 3
        continue

    print(i)
else:
    # Runs only if loop was NOT exited via break
    print("Loop completed without break")
```

## Problem 5: List Comprehensions

```python
# Problem 5: List Comprehensions
# Creates squares of even numbers from 1 to 20 using a list comprehension.

# Generate integers 1 to 20
integers = list(range(1, 21))

# Comprehension: square of n, but only when n is even
even_squares = [n ** 2 for n in integers if n % 2 == 0]

print("Squares of even numbers from 1 to 20:")
print(even_squares)
```

## Problem 6: Data Structure Manipulations

```python
# Problem 6: Data Structure Manipulations
# Demonstrates list → tuple → set and a dictionary of fruit colors.

# 1. Create a list of five fruits
fruits_list = ["apple", "banana", "cherry", "mango", "grape"]
print("List of fruits:", fruits_list)

# 2. Convert list to a tuple (immutable sequence)
fruits_tuple = tuple(fruits_list)
print("Tuple of fruits:", fruits_tuple)

# 3. Create a set from the list (unique, unordered) and add a new fruit
fruits_set = set(fruits_list)
fruits_set.add("kiwi")          # Add a sixth fruit to the set
print("Set of fruits (with kiwi added):", fruits_set)

# 4. Dictionary mapping fruit names to their colors
fruit_colors = {
    "apple":  "red",
    "banana": "yellow",
    "cherry": "dark red",
    "mango":  "orange",
    "grape":  "purple"
}
print("Dictionary of fruit colors:")
for fruit, color in fruit_colors.items():
    print(f"  {fruit}: {color}")
```

## Problem 7: Dictionary and Looping

```python
# Problem 7: Dictionary and Looping
# Collects three students' names and grades, then reports pass/fail.

student_grades = {}  # Dictionary: name → grade

# Collect data for three students
for i in range(3):
    name  = input(f"Enter name of student {i + 1}: ")
    grade = int(input(f"Enter grade for {name} (0-100): "))
    student_grades[name] = grade

print("Student Results:")
for student, score in student_grades.items():
    result = "Passed" if score >= 60 else "Failed" # A grade of 60 or above is considered passing
    print(f"  {student}: {score} → {result}")
```

## Problem 8: Defensive Programming with assert

```python
# Problem 8: Defensive Programming with assert
# divide_numbers uses assert to guard against division by zero.

def divide_numbers(a, b):
    """Divide a by b. Raises AssertionError if b is zero."""
    assert b != 0, "Divisor b must not be zero!"
    return a / b


# Demonstrate normal operation
result = divide_numbers(323, 123)
print(f"323 / 123 = {result:.2f}")

result = divide_numbers(71, 31)
print(f"71 / 31 = {result:.4f}")

# Demonstrate assertion failure (divide by zero)
try:
    divide_numbers(5, 0)
except AssertionError as error:
    print(f"AssertionError caught: {error}")
```

## Problem 9: Using pass in Conditionals

```python
# Problem 9: Using pass in Conditionals
# Checks if a character is a vowel, consonant (pass), or invalid input.

char = input("Enter a single character: ")

vowels = "aeiouAEIOU"

if len(char) != 1 or not char.isalpha():
    # Input is not a single alphabetic character
    print("Invalid input")
elif char in vowels:
    print("Vowel")
else:
    # Character is a consonant — intentionally do nothing
    pass
```

## Problem 10: Nested Loops and Pattern Printing

```python
# Problem 10: Nested Loops and Pattern Printing
# Prints a right-triangle star pattern with 5 rows.

rows = 5

# Outer loop controls the row number (1 through 5)
for row in range(1, rows + 1):
    # Inner loop prints 'row' number of stars on each line
    for _ in range(row):
        print("*", end="")
    print()  # Move to the next line after each row
```

```python
# Problem 11: Dictionary Comprehension
# Creates a dictionary of numbers 1-10 mapped to their cubes.

# Dictionary comprehension: {key: value for ...}
cubes_dict = {n: n ** 3 for n in range(1, 11)}

print("Number → Cube dictionary:")
for number, cube in cubes_dict.items():
    print(f"  {number}^3 = {cube}")
```

## Problem 12: Set Operations

```python
# Problem 12: Set Operations
# Demonstrates union, intersection, difference, and symmetric difference.

set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print(f"Set A: {set_a}")
print(f"Set B: {set_b}")

# Union — all elements from both sets (no duplicates)
print(f"Union (A | B):                {set_a | set_b}")

# Intersection — elements common to both sets
print(f"Intersection (A & B):         {set_a & set_b}")

# Difference — elements in A but not in B
print(f"Difference (A - B):           {set_a - set_b}")

# Symmetric Difference — elements in either set, but not both
print(f"Symmetric Difference (A ^ B): {set_a ^ set_b}")
```

## Problem 13: Filter with List Comprehensions

```python
# Problem 13: Filter with List Comprehensions
# Filters a list to keep only numbers divisible by 3.

numbers = [1, 3, 7, 9, 12, 14, 15, 18, 20, 21, 25, 27, 30]

# Comprehension: keep n only when it divides evenly by 3
divisible_by_3 = [n for n in numbers if n % 3 == 0]

print("Original list:", numbers)
print("Numbers divisible by 3:", divisible_by_3)
```

## Problem 14: Simple Menu-Driven Program

```python
# Problem 14: Simple Menu-Driven Program
# Allows the user to perform arithmetic operations in a loop until they exit.

# --- Operation functions ---
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Division by zero is not allowed."
    return x / y


# --- Main menu loop ---
while True:
    print("===== Calculator Menu =====")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ").strip()

    if choice == "5":
        print("Exiting calculator. Goodbye!")
        break  # Exit the loop

    if choice not in ("1", "2", "3", "4"):
        print("Invalid choice. Please enter a number between 1 and 5.")
        continue  # Ask again

    # Get the two numbers from the user
    a = float(input("Enter first number:  "))
    b = float(input("Enter second number: "))

    # Dispatch to the correct function based on choice
    if choice == "1":
        print(f"Result: {a} + {b} = {add(a, b)}")
    elif choice == "2":
        print(f"Result: {a} - {b} = {subtract(a, b)}")
    elif choice == "3":
        print(f"Result: {a} × {b} = {multiply(a, b)}")
    elif choice == "4":
        print(f"Result: {a} ÷ {b} = {divide(a, b)}")
```
