```python
filt = int(input("Enter 1 if you like to print even numbers, else enter 0: "))
for num in range(1,8):
  if filt:
    if num % 2 == 0:
      print(f"Even number: {num}")
  else:
    if num % 2 != 0:
      print(f"Odd number: {num}")
```

```python
arr_1 = []
arr_2 = [1,2,3]

arr_1 = arr_2 # a and b points to same list object
print(arr_1,arr_2)

arr_1 = arr_2[:] # a and b points to different list objects with same content
print(arr_1,arr_2)

```

```python
server = ['server1', 'server2', 'server3']

server.append('server4') # add at the end of the list
server.insert(1, 'server1.5') # insert at a position in this case at index 1

print(server)

server.remove('server1.5') # remove an element from the list
print(server)
server.pop() # remove the last element from the list
print(server)
```

```python
# List comprehension
squared_list = [num**2 for num in range(0,10)]
print(squared_list)
```

```python
# Copy lists
copy_list = squared_list.copy() # safe copy using slicing
print(copy_list)

squared_list.append(100)

print(squared_list)
print(copy_list)
```

```python
print(squared_list)
print(copy_list)
```

```python
# Copy lists
copy_list = squared_list[:] # shallow copy using slicing
print(copy_list)

squared_list.append(100)

print(squared_list)
print(copy_list)
```

```python
copy_list.sort(reverse=True)
print(copy_list)
```

```python
# Tuple
# Ordered, immutable, allows duplicates
my_tuple = (1,2,3,4,5)
print(my_tuple)
single_item_tuple = (1,) # to create a single item tuple, comma is necessary
print(single_item_tuple)my_tuple[0] = 10 # this will throw an error since tuples are immutable
```

```python
single_item_tuple = (1) # this is not tuple, it's an integer
print(type(single_item_tuple))

single_item_tuple = (1,) # to create a single item tuple, comma is necessary
print(type(single_item_tuple))

print(single_item_tuple)

print(len(single_item_tuple))
```

```python
t1 = (1,2,3)
t2 = (4,5,6)
print(t1 + t2)  # concatenation
print(t1 * 3)   # repetition

print(3 in t2)
```

```python
# in works for list, tuple, string, set, dictionary (checks in keys) etc.

print(3 in [1,2,3,4,5])
print(4 in (1,2,3))
print('a' in 'apple')
print('a' in {'a':1, 'b':2})
```

```python
# Tuple Packing
user_data = 'Danish', 25, 'University of Maryland' # this is tuple packing
print(user_data)

# Tuple Unpacking
name, age, university = user_data
print(name)
print(age)
print(university)
```

```python
# Sets Basics

# Unordered collection of unique elements
numbers_set = {1, 2, 3, 4, 5}
print(type(numbers_set))

numbers_set_with_duplicates = {1, 2, 3, 4, 5, 5, 4, 3, 2, 1}
print(numbers_set_with_duplicates)
```
