```python
for i in range(5):  # 5 is not included
  print(i) 
```

```python
for i in range(0,5):  # 5 is not included
  print(i) 
```

```python
for i in range(5,0,-1):  # 0 is not included, step is -1 to count down
  print(i) 
```

```python
# Step cannot be a float
for i in range(0,5,0.5):  # 0 is not included, step is -1 to count down
  print(i) 
```

```python

# Step cannot be a zero
for i in range(0,5,0):  # 0 is not included, step is -1 to count down
  print(i) 
```

```python
iter = 0
while True:
  print(iter)
  iter += 1
  if iter == 5:
    break
```

```python
temperature = 60
while temperature < 70:
  print(f"Temperature is too low: {temperature}")
  temperature += 1

print(f"Temperature is now acceptable: {temperature}")
```
