```python
def split_even_odd(nums):
  even = []
  odd = []
  for n in nums:
    if n % 2 == 0:
      even.append(n)
    else:
      odd.append(n)
  return even, odd
```

```python
even , odd = split_even_odd([1, 2, 3, 4, 5, 6])
print("Even numbers:", even)
print("Odd numbers:", odd)
```

```python
def min_max(nums):
  return min(nums), max(nums)
```

```python
min_number, max_number = min_max([3, 1, 4, 1, 5, 9])
print("Min:", min_number)
print("Max:", max_number)
```
