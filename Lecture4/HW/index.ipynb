```python
# Problem 1 – Basic Class Design
# Create	a	class	Person	with	attributes	name	(string)	and	age	(integer).
# Add	method	greet()	that	prints:	Hello,	my	name	is	[name]	and	I	am	[age]	years	old.
# Instantiate	two	objects	and	call	greet()
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  
  def greet(self):
    print(f'Hi! My name is {self.name} & I am {self.age} years old')
```

```python
# Problem 4 – Encapsulation with Validation
# Create	class	BankAccount	with	private	attribute	_balance.
# Add	deposit(amount)	and	withdraw(amount).
# Only	allow	positive	deposits.
# Prevent	overdraft.
# Raise	ValueError	when	invalid.
# Add	read-only	@property	balance.
```

```python
class BankAccount:
  def __init__(self, balance):
    if type(balance) is not int:
      raise ValueError("Balance must be an integer")
    else:
      self.__balance = balance

  @property
  def balance(self):
    return self.__balance
  
  def deposit(self, amount):
    if type(amount) is not int and type(amount) is not float:
      raise ValueError("Amount must be an integer")
    if amount < 0:
      raise ValueError('Invalid transaction')
    
    self.__balance += amount

  def withdraw(self, amount):
    print(type(amount))
    if type(amount) is not int and type(amount) is not float:
      raise ValueError("Amount must be an integer")
    if amount < 0:
      raise ValueError('Invalid transaction, please enter a positive value')
    
    if amount > self.__balance:
      raise ValueError('Insufficient Balance')
    
    self.__balance -= amount
```

```python
my_account = BankAccount(20000)
```

```python
my_account.withdraw(100)
```

```python
print('My balance is', my_account.balance)
```

```python
my_account.deposit(10000)
```

```python
print('My balance is', my_account.balance)
```

```python
my_account.deposit(-10000)
```

```python
my_account.withdraw(-10000)
```

```python
my_account.withdraw(22210000)
```

```python
# Problem 7 – Encapsulation with @property
# Create	class	Product	with	private	_price.
# Add	@property	getter	and	setter.
# Reject	negative	price	and	raise	ValueError.
# Demonstrate	valid	and	invalid	cases.
```

```python
class Product:
  def __init__(self, name, price):
    self.__name = name
    self.__price = price

  @property
  def name(self):
    return self.__name
  
  @property
  def price(self):
    return self.__price
  
  @price.setter
  def price(self, price):
    if price < 0:
      raise ValueError('Come on man! enter positive price')
    
    self.__price = price
```

```python
item = Product('iPhone', 1000)
```

```python
# Problem 8 – __repr__ for Debugging
# Modify	Product	to	include	__repr__	returning:
# Product(name=...,	price=...)
# Demonstrate	printing	the	object
```

```python
class Product:
  def __init__(self, name, price):
    self.__name = name
    self.__price = price

  @property
  def name(self):
    return self.__name
  
  @property
  def price(self):
    return self.__price
  
  @price.setter
  def price(self, price):
    if price < 0:
      raise ValueError('Come on man! enter positive price')
    
    self.__price = price

  def __repr__(self):
    return f"Product(name='{self.name}', price={self.price})"
```

```python
# Problem 9 – Production-Oriented Class
# Create	class	LibraryBook	with	title,	author,	year,	_is_checked_out.
# Add	methods:	check_out(),	return_book(),	status(),	__repr__().
# Validate	year	must	be	<=	current	year.
# Raise	ValueError	if	invalid.
# Demonstrate	full	workflow.
```

```python
class LibraryBook:
  def __init__(self, title, author, year, is_checked_out=False):
    if year > datetime.now().year:
      raise ValueError('Year should not be in the future')
    self.title = title
    self.author = author
    self.year = year
    self._is_checked_out = is_checked_out

  def checkout(self):
    if self._is_checked_out is True:
      print(f'{self.title} is not available')
    else:
      self._is_checked_out = True
      print(f'{self.title} is checked out. Thanks')

  def return_book(self):
    if self._is_checked_out is False:
      print(f'{self.title} is already returned')
    else:
      self._is_checked_out = False
      print(f'{self.title} is returned. Thanks')

  def status(self):
    status = 'Unavailable' if self._is_checked_out else 'Available'
    print(f'{self.title} by {self.author} is {status}')

  def __repr__(self):
    return f'LibraryBook(title={self.title}, author={self.author}, year={self.year}, is_checked_out={self._is_checked_out})'
```

```python
book = LibraryBook('Harry Potter', 'JK Rowling', 1997)
```

```python
# Problem 10 – Bonus (Inheritance)
# Create	base	class	Animal	with	make_sound().
# Subclass	Cat	overriding	make_sound().
# Demonstrate	polymorphism.
# (Optional	– 5	bonus	points)
```

```python
from abc import ABC, abstractmethod
class Animal(ABC):
  @abstractmethod
  def make_sound(self, sound: str) -> None:
    ...

class Cat(Animal):
  def make_sound(self) -> None:
    print(f'Cat goes Meow')


def lets_roll(animal: Animal) -> None:
  animal.make_sound()
```

```python
cat = Cat()
```

```python
lets_roll(cat)
```

```python
class Dog(Animal):
  def make_sound(self) -> None:
    print(f'Dog is barking')

class Cow(Animal):
  def make_sound(self) -> None:
    print(f'Cow goes Moooooow')
```

```python
dog = Dog()
cow = Cow()
```
